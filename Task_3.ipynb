```python
import numpy as np
import random
# n_lost - Количество потярянных пактеов
# n_rep - Количество повторов в методе Монте-Карло
# p_ref - Вероятность блокировки
# k - Количество концентраторов
# n_packets - Количество пакетов
# t - Текущее время
# lambda_day - Интесивность входного потока
# queue - текущий размер очереди
# m - Максимальный размер очереди
# mu - Интенсивность выходного потока
# t_free - время, когда прибор освободится
# V - Скорость передачи
# N - Средняя длина пакета


np.random.seed(100)


# в этой задаче мы ищем оптимальный размер буфера перебором
# так как максимальный размер очереди равен буфер - 1, то будем перебирать в программе предыдущей задачи
# по максимальному размеру, пока не найдём оптимальное значение.
# Так как днём интенсивность больше, будем перебирать для интенсивности днём


def cycle(buf):
    n_rep = 1000
    p_ref = 0
    V = 2400
    N = 1200
    for i in range(n_rep):
        k = 4
        n_packets = 1000
        t = 0
        queue = 0
        lambda_day = 4
        m = buf - 1
        mu = float(V) / N
        t_free = []
        for j in range(k):
            t_free.append(0)
        n_lost = 0
        for f in range(n_packets):
            t += np.random.exponential(scale=1/lambda_day)
            n = 0
            if queue > 0:
                for j in range(k):
                    while t_free[j] < t and queue > 0:
                        t_free[j] += np.random.exponential(scale=1/mu)
                        queue -= 1
            for h in range(k):
                if t_free[h] < t:
                    t_free[h] = t + np.random.exponential(scale=1/mu)
                    break
                else:
                    n += 1
            if n == k:
                if queue < m:
                    queue += 1
                else:
                    n_lost += 1
        p_ref += n_lost/n_packets

    p_ref /= n_rep
    return p_ref


buf = 5
res = cycle(buf)
while res > 0.0001:
    print('Вероятность блокировки: ' + str(res), ', буфер: ', buf)
    buf += 1
    res = cycle(buf)

print('Результат - вероятность блокировки: '+str(res), ', буфер: ', buf)
```
