```python
import numpy as np
import random

# k - Количество концентраторов
# lambda_d - Интесивность входного потока
# lambda_n - Интенсивность ночного потока
# m - Максимальный размер очереди
# mu - Интенсивность выходного потока
# p_d, p_n - Количество обслуженных заявок в единицу времени
# pd - Вероятность отказа днём
# pn - Вероятность отказа ночью
# V - Скорость передачи
# N - Средняя длина пакета

# в этой задаче мы вычисляем вероятность блокировки по специальной формуле 
# для многоканальной СМО с общей очередью
#   p = (p_x**(k+m)/((k**m)*f))*p0
#   p0 = (1 + po + po^2/2! + ... + (po^k)/(k)! + (1-(p_d/k)**m)/(1-p_d/k)))) ^ (-1)
#   p_x == p_n или p_x == p_d в зависимости от того, для какого времени суток рассчёт

k = 4
V = 2400
N = 1200
lambda_d = 4
lambda_n = 0.5
m = 4
mu = float(V) / N
p_d = lambda_d/mu
p_n = lambda_n/mu
p0 = 0
f = 1
for i in range(1, k+1):
    f *= i
    p0 += (p_d**i)/f
p0 = (p0 + 1 + (((p_d**(k+1))/(k*f)) * (1-(p_d/k)**m)/(1-p_d/k)))**(-1)
pd = (p_d**(k+m)/((k**m)*f))*p0

p0 = 0
f = 1
for i in range(1, k+1):
    f *= i
    p0 += (p_n**i)/f
p0 = (p0 + 1 + (((p_n**(k+1))/(k*f)) * (1-(p_n/k)**m)/(1-p_n/k)))**(-1)
pn = (p_n**(k+m)/((k**m)*f))*p0

print('Вероятность днём:', pd, ', ночью:', pn)
```

Вероятность днём: 0.00546448087431694 , ночью: 1.9341691114032334e-09
