```python
import numpy as np
import random
# k - Количество концентраторов
# n_r - Количество повторов в методе Монте-Карло
# t - Текущее время
# mu - Интенсивность выходного потока
# t_free время освобождения прибора
# queue - очередь
# m - Максимальный размер очереди
# p_r - Вероятность блокировки
# n_l - Количество потерянных пакетов
# V - Скорость передачи
# N - Средняя длина пакета


# в этом задании мы симулируем работу комутаторов в реальном времени.
# для этого мы будем посылать пакеты в систему комутаторов через неравные промежутки времени
# и пронаблюдаем реакцию системы
# таким образом можно найти количество пакетов, потерянных в системе.
# найдя кол-во утерянных пакетов можно найти вероятность отказа.
# вероятность ночью равна нулю, так как интенсивность ночью очень мала, 
# в связи с этим пакеты не успевают потеряться
def monte_carlo(lambda_, n_packets, n_r):
    np.random.seed(100)
    p_r = 0
    V = 2400
    N = 1200
    for i in range(n_r):
        k = 4
        t = 0
        queue = 0
        m = 4
        mu = float(V) / N
        t_free = []
        for j in range(k):
            t_free.append(0)
        n_l = 0
        for f in range(n_packets):
            t += np.random.exponential(scale=1/lambda_)
            n = 0
            if queue > 0:
                for j in range(k):
                    while t_free[j] < t and queue > 0:
                        t_free[j] += np.random.exponential(scale=1/mu)
                        queue -= 1
            for h in range(k):
                if t_free[h] < t:
                    t_free[h] = t + np.random.exponential(scale=1/mu)
                    break
                else:
                    n += 1
            if n == k:
                if queue < m:
                    queue += 1
                else:
                    n_l += 1
        p_r += n_l/n_packets
    return p_r



n_r = 100
p_ref1 = monte_carlo(4, 230400, n_r) / n_r       # 230400 == 16h * 3600s * 4
p_ref2 = monte_carlo(0.5, 14400, n_r) / n_r      # 14400 == 8h * 3600s * 0.5
print('Вероятность отказа днём: ' + str(p_ref1) + ', ночью: ' + str(p_ref2))
```
